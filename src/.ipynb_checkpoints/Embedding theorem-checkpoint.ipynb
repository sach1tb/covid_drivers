Takens' theorem


## Rossler attractor

\begin{equation}
\begin{split}
\frac{dx}{dt}&=-y-z\\
\frac{dy}{dt}&=x+ay\\
\frac{dz}{dt}&=b+z(x-c)
\end{split}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def rossler(Y,t,a, b,c):
    x,y,z = Y
    dYdt=[-y-z, x+a*y, b+z*(x-c)]
    return dYdt

y0 = [-8, 0,1]
a=0.1
b=0.1
c=14
t = np.arange(0, 60, 0.02)
sol = odeint(rossler, y0, t, args=(a,b,c))

ax = plt.figure().add_subplot(projection='3d')
ax.plot(sol[:, 0],sol[:, 1], sol[:,2], 'b')

ax = plt.figure().add_subplot(projection='3d')
tau=35
T=sol.shape[0]
# can use np.roll but it produces unsightly jumps in in the initial points
ax.plot(sol[0:T-2*tau, 0],sol[tau:T-tau, 0], sol[2*tau:T,0], 'b')

plt.show()
```
